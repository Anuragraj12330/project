importlibrary

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Create Sample Dataset

```python
data = {
    "Month": pd.date_range(start="2024-01-01", periods=12, freq="M"),
    "Electricity_kWh": [3200, 3000, 3400, 3800, 4500, 4800, 4700, 4600, 4100, 3600, 3300, 3100],
    "Water_KL": [120, 115, 118, 130, 145, 150, 148, 147, 135, 125, 122, 119],
    "Cost_Rs": [24000, 22500, 25000, 28000, 34000, 36500, 36000, 35500, 31000, 27000, 24500, 23000]
}

df = pd.DataFrame(data)
```

Display Datase


```python
print("Dataset:\n", df)
```

Descriptive Statistics

```python
print("\nDescriptive Statistics:\n", df.describe())
```

Variance

```python
print("\nVariance:")
print(df[["Electricity_kWh", "Water_KL", "Cost_Rs"]].var())
```

Line Plot - Electricity Usage

```python
plt.figure()
plt.plot(df["Month"], df["Electricity_kWh"], marker='o')
plt.xlabel("Month")
plt.ylabel("Electricity Consumption (kWh)")
plt.title("Monthly Electricity Consumption Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

 Line Plot - Water Usage

```python
plt.figure()
plt.plot(df["Month"], df["Water_KL"], marker='o')
plt.xlabel("Month")
plt.ylabel("Water Consumption (KL)")
plt.title("Monthly Water Consumption Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Line Plot - Cost Trend

```python
plt.figure()
plt.plot(df["Month"], df["Cost_Rs"], marker='o')
plt.xlabel("Month")
plt.ylabel("Cost (Rs)")
plt.title("Monthly Utility Cost Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
